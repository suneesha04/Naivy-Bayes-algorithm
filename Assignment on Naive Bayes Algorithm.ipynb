```python
nb_test['workclass']=nb_test['workclass'].astype('category')
nb_test['education']=nb_test['education'].astype('category')
nb_test['maritalstatus']=nb_test['maritalstatus'].astype('category')
nb_test['occupation']=nb_test['occupation'].astype('category')
nb_test['relationship']=nb_test['relationship'].astype('category')
nb_test['race']=nb_test['race'].astype('category')
nb_test['sex']=nb_test['sex'].astype('category')
nb_test['native']=nb_test['native'].astype('category')
nb_test['Salary']=nb_test['Salary'].astype('category')
```
